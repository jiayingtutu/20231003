```python
s=input("請輸入五個成績並以空格分開")
s = s.split(" ")
s.sort()
del s[0]
del s[-1]
total = (float(s[0]) + float(s[1]) + float(s[2]))
avg = total / 3
print(avg)
```

```python
a = float(input("請輸入一個數字"))
b = float(input("請輸入另一個數字"))
c = input("請輸入一個運算符號")
if c=="+":
  x=a+b
elif c=="-":
  x=a-b
elif c=="*":
  x=a*b
else:
  x=a/b

print(x)
```

```python
y = int(input("請輸入年分"))
if y > 1582:
  if y%400==0:
    print("是閏年")
  elif y%4==0 and y%100!=0:
    print("是閏年")
  else:
    print("是平年")
elif y%4==0:
  print("是閏年")
else:
  print("是平年")
```
