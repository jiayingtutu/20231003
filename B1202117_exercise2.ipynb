```python
h = float(input("請輸入身高"))
w = float(input("請輸入體重"))
H = h/100
H*=H
w/=H
print("BMI:",w)
if w<18.5:
  print("體位:過輕")
elif w<24:
  print("體位:適中")
elif w<27:
  print("體位:過重")
else:
  print("體位:肥胖")
```
